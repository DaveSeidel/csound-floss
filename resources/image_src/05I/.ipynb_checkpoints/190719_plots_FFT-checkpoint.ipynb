```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
figure(figsize=(10,5))
plot(hamming(1024),c='g',label='Hamming')
plot(hanning(1024),c='r',label='von-Hann (Hanning)')
xlim(0,1024)
ylim(0,1.05)
legend()
#suptitle('Hamming and Hanning window (512 samples)')
```

```python
f,a = subplots(figsize=(15,5))
plot(arange(0,512),hanning(512))
plot(arange(128,512+128),hanning(512))
plot(arange(256,512+256),hanning(512))
plot(arange(384,512+384),hanning(512))
xlim(0,512+384)
ylim(0,1.05)
xticks([0,128,256,384,512,512+128,512+256,512+384])
xlabel('samples')
#suptitle('Four overlapping Hanning windows (each of size=512 samples)')
```
