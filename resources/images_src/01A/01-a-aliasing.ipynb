```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig, ax = plt.subplots(1,1, dpi=100)
x = np.linspace(0,50,500)
x_samps = np.linspace(0,48,10)

ax.set(xticks=[], yticks=[], xlim=(0,50))
ax.set_xlabel('Time', x=1)
ax.set_ylabel('Amplitude')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_position('zero')
ax.plot(x,np.sin(x), c='b')
ax.plot(x, -np.sin(x/5.6), c='r')
ax.bar(x_samps,np.sin(x_samps),width=0.3, color='k')
ax.scatter(x_samps,np.sin(x_samps), c='r')

fig.savefig('01A_aliasing_1b.png', bbox_inches='tight')
```

```python
fig, (ax1, ax2) = plt.subplots(2,1, dpi=120)
fig.suptitle('Sample Rate = 40 kHz (40000 Samples / Second)', x=0.3)
x = np.linspace(0,10,100)
x_samps = np.linspace(0,np.pi*3,7)
# top
ax1.set(xticks=[], yticks=[], xlim=(0,10.5))
ax1.set_xlabel('Time', x=1)
ax1.set_ylabel('10 kHz\nSine Wave', rotation=0, labelpad=30)
ax1.spines['top'].set_visible(False)
ax1.spines['right'].set_visible(False)
ax1.spines['bottom'].set_position('zero')
ax1.plot(x,np.sin(x), c='b')
ax1.bar(x_samps,np.sin(x_samps),width=0.05, color='k')
ax1.scatter(x_samps, np.sin(x_samps), c='r')
# bottom
ax2.set(xticks=[], yticks=[], xlim=(0,10.5))
ax2.set_xlabel('Time', x=1)
ax2.set_ylabel('30 kHz\nSine (blue)\n\n10 KHz aliased\nSine (red)', rotation=0, labelpad=40, y=0.2)
ax2.spines['top'].set_visible(False)
ax2.spines['right'].set_visible(False)
ax2.spines['bottom'].set_position('zero')
ax2.plot(x,-np.sin(3*x), c='b')
ax2.bar(x_samps,np.sin(x_samps),width=0.05, color='k')
ax2.scatter(x_samps, np.sin(x_samps), c='r')
ax2.plot(x,np.sin(x), c='r')

fig.savefig('01A_aliasing_2b.png', bbox_inches='tight')
```
